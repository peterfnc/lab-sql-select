```python
import os

os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="/Users/ferencipeter/Ironhack Data Analytics.json"

from google.cloud import bigquery

client = bigquery.Client()
```

```python
query = '''
SELECT 
  auth.au_id,
  auth.au_lname,
  auth.au_fname,
  tit.title,
  pubs.pub_name
FROM 
  `ironhack-data-analytics.publications.authors` AS auth
JOIN
  `ironhack-data-analytics.publications.titleauthor` AS titau
  ON titau.au_id = auth.au_id
JOIN
  `ironhack-data-analytics.publications.titles` AS tit
  ON tit.title_id = titau.title_id
JOIN
  `ironhack-data-analytics.publications.publishers` AS pubs
  ON tit.pub_id = pubs.pub_id
LIMIT 1000
'''

query_job = client.query(query)

df=query_job.to_dataframe()
df.head(30)
```

```python
query2 = '''
SELECT 
  auth.au_id,
  auth.au_lname,
  auth.au_fname,
  pubs.pub_name,
  COUNT(tit.title) AS title_count
FROM 
  `ironhack-data-analytics.publications.authors` AS auth
JOIN
  `ironhack-data-analytics.publications.titleauthor` AS titau
  ON titau.au_id = auth.au_id
JOIN
  `ironhack-data-analytics.publications.titles` AS tit
  ON tit.title_id = titau.title_id
JOIN
  `ironhack-data-analytics.publications.publishers` AS pubs
  ON tit.pub_id = pubs.pub_id
GROUP BY
  1,2,3,4
ORDER BY
  title_count DESC
LIMIT 1000
'''

query_job = client.query(query2)

df=query_job.to_dataframe()
df.head(30)
```

```python
query3 = '''
SELECT 
  auth.au_id,
  auth.au_lname,
  auth.au_fname,
  SUM(sale.qty) AS titles_sold
FROM 
  `ironhack-data-analytics.publications.authors` AS auth
JOIN
  `ironhack-data-analytics.publications.titleauthor` AS titau
  ON titau.au_id = auth.au_id
JOIN
  `ironhack-data-analytics.publications.titles` AS tit
  ON tit.title_id = titau.title_id
JOIN
  `ironhack-data-analytics.publications.publishers` AS pubs
  ON tit.pub_id = pubs.pub_id
JOIN
  `ironhack-data-analytics.publications.sales` AS sale
  ON tit.title_id = sale.title_id
GROUP BY
  1,2,3
ORDER BY
  titles_sold DESC
LIMIT 3
'''

query_job = client.query(query3)

df=query_job.to_dataframe()
df.head(30)
```

```python
query4 = '''
SELECT 
  auth.au_id,
  auth.au_lname,
  auth.au_fname,
  SUM(sale.qty) AS titles_sold
FROM 
  `ironhack-data-analytics.publications.authors` AS auth
LEFT JOIN
  `ironhack-data-analytics.publications.titleauthor` AS titau
  ON auth.au_id = titau.au_id
LEFT JOIN
  `ironhack-data-analytics.publications.titles` AS tit
  ON tit.title_id = titau.title_id
LEFT JOIN
  `ironhack-data-analytics.publications.publishers` AS pubs
  ON tit.pub_id = pubs.pub_id
LEFT JOIN
  `ironhack-data-analytics.publications.sales` AS sale
  ON tit.title_id = sale.title_id
GROUP BY
  1,2,3
ORDER BY
  titles_sold DESC
LIMIT 100
'''

query_job = client.query(query4)

df=query_job.to_dataframe()
df.head(30)
```
